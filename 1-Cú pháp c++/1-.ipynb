# các loại kiểu dữ liệu

(đoán mò) được quy định là mỗi loại máy tính định nghĩa nó khác nhau

```java
// kiểu ký tự
char        : ít nhất được quy định dài 8 bit,  có thể dài hơn
char16_t    : ít nhất được quy định dài 16 bit, có thể dài hơn, chắc chắn dài hơn char
char32_t    : ít nhất được quy định dài 32 bit, có thể dài hơn, chắc chắn dài hơn char16_t
wchar_t     : biểu diễn nhiều loại ký tự nhất
```

```java
// kiểu số nguyên có dấu âm/ dương
signed char             : ít nhất được quy định dài 8 bit,  có thể dài hơn 
signed short int        : ít nhất được quy định dài 16 bit, có thể dài hơn, chắc chắn dài hơn char
signed int              : ít nhất được quy định dài 16 bit, có thể dài hơn, chắc chắn dài hơn short
signed long int         : ít nhất được quy định dài 32 bit, có thể dài hơn, chắc chắn dài hơn int
signed long long int    : ít nhất được quy định dài 64 bit, có thể dài hơn, chắc chắn dài hơn long
```

```java
// kiểu số nguyên không có dấu âm
unsigned char           : ít nhất được quy định dài 8 bit,  có thể dài hơn 
unsigned short int      : ít nhất được quy định dài 16 bit, có thể dài hơn, chắc chắn dài hơn char
unsigned int            : ít nhất được quy định dài 16 bit, có thể dài hơn, chắc chắn dài hơn short
unsigned long int       : ít nhất được quy định dài 32 bit, có thể dài hơn, chắc chắn dài hơn int
unsigned long long int  : ít nhất được quy định dài 64 bit, có thể dài hơn, chắc chắn dài hơn long
```

```java
// kiểu số thực
float 
double      : độ chính xác được quy định cao hơn float
long double : độ chính xác được quy định cao hơn double
```

# xác định kiểu bằng auto và decltype

```java
// auto a1 = a2; 
// a2 đã khởi tạo, a1 chưa khởi tạo
// -> khởi tạo a1 có kiểu của a2
// -> khởi tạo giá trị của a1 là a2
```

```java
// decltype(a2) a1;
// a2 đã khởi tạo, a1 chưa khởi tạo
// -> khởi tạo a1 có kiểu của a2
```
