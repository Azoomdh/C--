# overload

```python
// overloading functions
#include <iostream>
using namespace std;

int operate (int a, int b) {
    return (a*b);
}

double operate (double a, double b) {
    return (a/b);
}

int main () {
    int x=5,y=2;
    double n=5.0,m=2.0;
    cout << operate (x,y) << '\n';
    cout << operate (n,m) << '\n';
    return 0;
}
```

# template

* ## template là class

```python
// function template
#include <iostream>
using namespace std;

template <class T>
T sum (T a, T b) {
    T result;
    result = a + b;
    return result;
}

int main () {
    int i=5, j=6;
    double f=2.0, g=0.5;

    // tự suy luận kiểu
    cout << sum<int>(i,j) << " " << sum(i, j) << '\n';
    cout << sum<double>(f,g) << " " << sum(f, g)<< '\n';
    
    return 0;
}
```

```python
// function templates
#include <iostream>
using namespace std;

template <class T, class U>
bool are_equal (T a, U b) {
    return (a==b);
}

int main () {
    // tự suy luận kiểu
    if (are_equal(10,10.0) && are_equal<int,double>(10,10.0) ) {
        cout << "x and y are equal\n";
    }
    else {
        cout << "x and y are not equal\n";
    }
    return 0;
}
```

* ## template là int không phải class

```python
// template arguments
#include <iostream>
using namespace std;

template <class T, int N>
T fixed_multiply (T val) {
    return val * N;
}

int main() {
    std::cout << fixed_multiply<int,2>(10) << '\n';
    std::cout << fixed_multiply<int,3>(10) << '\n';
}
```
