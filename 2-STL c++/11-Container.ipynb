# Container

```python
// danh sách hàm của Container
Container a , b;

a == b;
a < b

a.begin()   = con trỏ tới phần tử đầu tiên 
a.end()     = con trỏ tới phần tử cuối

a.size()
a.empty()

a.front()   = phần tử đầu tiên
a.back()    = phần tử cuối

a[5]        = phần tử thứ 5

a.insert()
a.push_back()
a.push_front()

a.erase()
a.pop_back()
a.pop_front()
```
