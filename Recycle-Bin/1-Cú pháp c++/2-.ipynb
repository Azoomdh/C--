# các lệnh thông thường

* ## if else

````python
if (x > 0) {
    cout << "x is positive";
}
else if (x < 0) {
    cout << "x is negative";
}
else {
    cout << "x is 0";
}
````

* ## while

````python
int n = 10;

while (n>0) {
    cout << n << ", ";
    --n;
}
````

* ## do-while

````python
string str;
do {
    cout << "Enter text: ";
    getline (cin,str);
    cout << "You entered: " << str << '\n';
} 
while (str != "goodbye");
````

* ## hàm for

````python
for (int n=10; n>0; n--) {
    cout << n << ", ";
}
````

# lệnh nhảy

* ## break : ngừng hẳn vòng lặp for/while chứa nó

````python
// break loop example
#include <iostream>
using namespace std;

int main () {
    for (int n=10; n>0; n--) {
        cout << n << ", ";
        if (n==3) {
            cout << "countdown aborted!";
            break;
        }
    }
}
````

* ## continue : bỏ qua các dòng code bên dưới tại vòng lặp đó

````python
// continue loop example
#include <iostream>
using namespace std;

int main () {
    for (int n=10; n>0; n--) {
        if (n==5) { 
            continue; // bỏ qua code bên dưới khi n== 5
        }
        cout << n << ", ";
    }
    cout << "liftoff!\n";
}
````

* ## lệnh goto : nhảy

````python
// goto loop example
#include <iostream>
using namespace std;

int main () {
    int n=10;
    mylabel:
    cout << n << ", ";
    n--;
    if (n>0) {
        goto mylabel;
    }
    cout << "liftoff!\n";
}
````

# switch case

````python
// thử vào file Test.cpp

#include<iostream>

using namespace std;

int main(){
    int day = 1;
    switch (day) {
        case 1: {
            cout << "Monday";
            break;
        }
        case 2: {
            cout << "Tuesday";
            break;
        }
        case 3: {
            cout << "Wednesday";
            break;
        }
        case 4: {
            cout << "Thursday";
            break;
        }
        case 5: {
            cout << "Friday";
            break;
        }
        case 6: {
            cout << "Saturday";
            break;
        }
        case 7: {
            cout << "Sunday";
            break;
        }
        default : {
            cout << "invalid day";
            break;
        }
    }
}
````
