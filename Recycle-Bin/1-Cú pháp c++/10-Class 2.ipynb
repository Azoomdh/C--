## // không có virtual

```python
class Hinh {
    public :
        🟥 void printDetail(){
            cout << "day la hinh ???";
        }
};

class TamGiac1 : public Hinh {
    public : 
        void printDetail(){
            cout << "day la hinh tam giac" << endl ;
        }
}

class TamGiac2 : public Hinh {
    public :
        void printDetail(){
            cout << "day la hinh tam giac 2" << endl ;
        }
}

Hinh *h0;
TamGiac1 h1 = new TamGiac1();
TamGiac2 h2 = new TamGiac2();

h0 = &h1; 
h0->printDetail(); // day la hinh ???

h0 = &h2;
h0->printDetail(); // day la hinh ???
```

## // có virtual

```python
class Hinh {
    public :
        🟩 virtual void printDetail(){
            cout << "day la hinh ???";
        }
};

class TamGiac1: public Hinh {
    public : 
        void printDetail(){
            cout << "day la hinh tam giac 1" << endl ;
        }
}

class TamGiac2 : public Hinh {
    public :
        void printDetail(){
            cout << "day la hinh tam giac 2" << endl ;
        }
}

Hinh *h0;
TamGiac1 h1 = new TamGiac1();
TamGiac2 h2 = new TamGiac2();

h0 = &h1; 
h0->printDetail(); // day la hinh tam giac 1

h0 = &h2;
h0->printDetail(); // day la hinh tam giac 2
```

## //

```python
// h0 : con trỏ, định nghĩa sẽ trỏ tới 1 Hinh, hiện tại đang trỏ tới h1
// h1 : biến, định nghia là TamGiac, hiện tại đang là TamGiac
h0 = &h1;
h0->printDetail()
```

```python
// nếu Hinh không có virtual : 
    // h0->printDetail() sẽ bị đối xử theo định nghĩa của h0
    // h0 có định nghĩa là Hinh

// nếu Hinh có virtual
    // h0->printDetail() sẽ bị đối xử theo định nghĩa của &h1
    // &h1 là hình tam giác 1
```
