```python
// classes example
#include <iostream>
using namespace std;

class Rectangle {
    private :
        int width, height;
    public:
        void set_values(int, int);
        int area() {
            return width*height;
        }
};

// dấu ::
void Rectangle::set_values (int x, int y) {
    width = x;
    height = y;
}

int main () {
    Rectangle rect;
    rect.set_values (3,4);
    cout << "area: " << rect.area();
    return 0;
}
```

* # hàm khởi tạo

```python
// example: class constructor
#include <iostream>
using namespace std;

class Rectangle {
    private :
        int width, height;
    public:
        Rectangle (int,int);
        int area () {
            return (width*height);
        }
};

Rectangle::Rectangle (int a, int b) {
  width = a;
  height = b;
}

int main () {
    Rectangle rect (3,4);

    cout << rect.area() << endl;
}
```

* # Khởi tạo thành viên

```python
// 2 dòng dưới là tương tự nhau

Rectangle::Rectangle (int x, int y) { width=x; height=y; }

Rectangle::Rectangle (int x, int y) : width(x), height(y) { }
```
