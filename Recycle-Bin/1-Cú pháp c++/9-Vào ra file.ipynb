# mở tệp

````python
// chế độ
ios::in         = chế độ đọc
ios::out        = chế độ ghi
ios::ate        = mở tệp tin, đặt con trỏ ở cuối tệp
ios::trunc      = nếu tệp tin đã có sẵn thì sẽ bị xóa trước
ios::nocreate   = không tạo mới nếu tệp này không tồn tại
ios::noreplace  = chỉ mở khi tệp chưa tồn tại
ios::binary     = chế độ nhị phân
ios::text       = chế độ văn bản 
````

````python
// 🟥🟥🟥🟥 tệp tin mở được nhiều chế độ cùng lúc 🟥🟥🟥🟥
// Ví dụ, muốn mở một tệp tin abc.txt để đọc (ios::in) đồng thời với để ghi (ios::out) dưới chế độ văn bản (ios::text), ta khai báo như sau:
fstream myFile(“abc.txt”, ios::in|ios::out|ios::text);
````

# IO tệp văn bản

## ghi tệp văn bản

````python
// mở tập tin
ofstream fileOut('C://...//Downloads//...', ios::out);

// nếu không mở được thì thoát
if(!fileOut){
    cout << "khong mo duoc tep tin" << endl;
    exit(1);
}

// ghi dữ liệu vào tệp tin
do {
    string input;
    cin >> input;
    fileOut << input << "-";
} while(input != 'e' && (fileOut) );

// xuống cuối tệp tin
fileOut << endl;

// đóng tệp tin
fileOut.close();
````

## đọc tệp văn bản

````python
// mở tập tin
ifstream fileIn('C://...//Downloads//...', ios::in);

// nếu không mở được thì thoát
if( !fileIn ){
    cout << "không mở được tệp tin" << endl;
    exit(1);
}

// đọc dữ liệu từ tệp tin
while (fileIn){
    string output;
    fileIn >> output;
    cout << output;
}

// đóng tệp tin
fileIn.close();
````

# IO tệp nhị phân

## ghi đè toán tử cin/cout

````python
class Employee {
    // ...

    friend ostream& operator<< (ostream& output, const Employee& empl1){
        output << "id : " << empl1.id <<endl;
        output << "name : " << empl1.name << endl;

        return output;
    }

    friend istream& operator>> (istream& input, Employee& empl1){
        cout << "empl id : ";
        input >> empl1.id ;
        
        cout << "empl name : ";
        input >> empl1.name ;

        return input;
    }
}

int main(){
    Employee empl1;
    cin>>empl1;
    cout<<empl1;
}
````

## ghi tệp nhị phân

````python
// mở tập tin
fstream fileOut('C://...//Download//...', ios::out|ios::binary);

// nếu không mở được thì thoát
if(! fileOut){
    cout << "không mở được tệp tin" <<endl;
    exit(1);
}

// nhập các empl
for(int i=0; i<10; i++){
    Employee empl1;
    cin >> empl1;

    fileOut.write(
        reinterpret_cast<char*>(&empl1), 
        sizeof(Employee)
    );
}

// đóng tệp tin
fileOut.close()
````

## đọc tệp nhị phân

````python
// mở tệp tin
fstream fileIn('C://...//Download//...', ios::in|ios::binary);

// nếu không mở được thì thoát
if( !fileOut){
    cout << "không mở được tệp tin" << endl;
    exit(1);
}

// đọc tệp và in ra màn hình
while(fileIn) {
    Employee empl1;
    fileIn.read(
        reinterpret_cast<char*>(&empl1),
        sizeof(Employee)
    );
}

// đóng tệp
fileIn.close();
````
