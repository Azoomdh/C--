Các thành phần của STL

Các thành phần của STL là các tính năng được cung cấp bởi Thư viện mẫu chuẩn (STL) trong C++ có thể được phân loại thành 4 loại:

1.  Container
2.  Thuật toán
3.  Trình lặp
4.  Các hàm tử

Các thành phần này được thiết kế để đạt hiệu quả, linh hoạt và có thể tái sử dụng, khiến chúng trở thành một phần không thể thiếu của lập trình C++ hiện đại.

Container

[Container](https://www.geeksforgeeks.org/cpp/containers-cpp-stl/) là các cấu trúc dữ liệu được sử dụng để lưu trữ các đối tượng và dữ liệu theo yêu cầu. Mỗi container được triển khai dưới dạng một lớp mẫu, đồng thời chứa các phương thức để thực hiện các thao tác cơ bản trên đó. Mỗi container STL được định nghĩa bên trong tệp tiêu đề riêng của nó.

Container có thể được phân loại thành 4 loại:

1.  Container trình tự
2.  Bộ chuyển đổi container
3.  Container liên kết
4.  Container liên kết không có thứ tự

Nếu bạn muốn tìm hiểu sâu về STL và hiểu hết tiềm năng của nó, [Khóa học C++ đầy đủ](https://www.geeksforgeeks.org/courses/cpp-programming-basic-to-advanced?utm_campaign=221_cpp_standard_template_library_stl&utm_medium=gfgcontent_icp&utm_source=geeksforgeeks) của chúng tôi cung cấp hướng dẫn đầy đủ về cách làm chủ các container, trình lặp và thuật toán do STL cung cấp.

Container trình tự

Container trình tự lưu trữ dữ liệu theo cách tuyến tính. Chúng cũng được sử dụng để triển khai các bộ điều hợp container.

Có 5 container trình tự trong C++ STL:

1.  [Mảng](https://www.geeksforgeeks.org/cpp/stdarray-in-cpp/) : Mảng STL là một triển khai của một mảng không thể thay đổi kích thước khi biên dịch. Nó chứa nhiều phương thức cho các thao tác mảng thông thường.
2.  [Vector](https://www.geeksforgeeks.org/cpp/vector-in-cpp-stl/) : Vector STL có thể được định nghĩa là mảng có kích thước động, có thể tự động thay đổi kích thước khi thêm hoặc xóa các phần tử mới.
3.  [Deque](https://www.geeksforgeeks.org/cpp/deque-cpp-stl/) : Deque hay Hàng đợi hai đầu là các container tuần tự có tính năng mở rộng và thu hẹp ở cả hai đầu. Điều này có nghĩa là chúng ta có thể thêm và xóa dữ liệu vào và ra khỏi cả hai đầu.
4.  [Danh sách](https://www.geeksforgeeks.org/cpp/list-cpp-stl/) : Bộ chứa danh sách lưu trữ dữ liệu trong bộ nhớ không liền kề, không giống như các vectơ và chỉ cung cấp quyền truy cập tuần tự vào dữ liệu được lưu trữ. Về cơ bản, nó triển khai danh sách liên kết kép.
5.  [Danh sách chuyển tiếp](https://www.geeksforgeeks.org/cpp/forward-list-c-set-1-introduction-important-functions/) : Danh sách chuyển tiếp cũng lưu trữ dữ liệu theo cách tuần tự như danh sách, nhưng khác biệt là danh sách chuyển tiếp chỉ lưu trữ vị trí của các phần tử tiếp theo trong chuỗi. Nó thực hiện danh sách liên kết đơn.

Bộ chuyển đổi container

Bộ điều hợp container là loại container STL điều chỉnh các lớp container hiện có để phù hợp với nhu cầu hoặc yêu cầu cụ thể.

Có 3 bộ điều hợp container trong C++ STL:

1.  [Stack](https://www.geeksforgeeks.org/cpp/stack-in-cpp-stl/) : Stack STL tuân theo nguyên tắc "Vào Sau Ra Trước" (LIFO) về việc chèn và xóa phần tử. Hơn nữa, các thao tác này chỉ được thực hiện ở một đầu của stack.
2.  [Queue](https://www.geeksforgeeks.org/cpp/queue-cpp-stl/) : Hàng đợi STL tuân theo nguyên tắc First In First Out (FIFO), nghĩa là phần tử nào được chèn trước thì được xóa trước, và phần tử nào được chèn sau thì được xóa sau. Mặc định, hàng đợi STL sử dụng deque container.
3.  [Hàng đợi Ưu tiên](https://www.geeksforgeeks.org/cpp/priority-queue-in-cpp-stl/) : Hàng đợi Ưu tiên STL không tuân theo bất kỳ nguyên tắc FIFO hay LIFO nào, nhưng việc xóa phần tử được thực hiện dựa trên mức độ ưu tiên của nó. Vì vậy, phần tử có mức độ ưu tiên cao nhất (mặc định) luôn được xóa trước. Theo mặc định, nó sử dụng vector làm container cơ sở.

Container liên kết

Container liên kết là loại container lưu trữ các phần tử theo thứ tự được sắp xếp dựa trên khóa thay vì thứ tự chèn của chúng.

Có 4 loại container liên kết trong C++ STL:

1.  [Sets](https://www.geeksforgeeks.org/cpp/set-in-cpp-stl/) : STL Set là một loại container liên kết, trong đó mỗi phần tử phải là duy nhất vì giá trị của phần tử đó xác định phần tử đó. Theo mặc định, các giá trị được lưu trữ theo thứ tự tăng dần.
2.  [Bản đồ](https://www.geeksforgeeks.org/cpp/map-associative-containers-the-c-standard-template-library-stl/) : Bản đồ STL là các vùng chứa liên kết lưu trữ các phần tử dưới dạng cặp khóa-giá trị. Các khóa phải là duy nhất và vùng chứa được sắp xếp dựa trên giá trị của các khóa.
3.  [Multisets](https://www.geeksforgeeks.org/cpp/multiset-in-cpp-stl/) : STL Multiset tương tự như container set ngoại trừ việc nó có thể lưu trữ các giá trị trùng lặp.
4.  [Multimaps](https://www.geeksforgeeks.org/cpp/multimap-associative-containers-the-c-standard-template-library-stl/) : STL Multimap tương tự như một vùng chứa bản đồ nhưng cho phép nhiều giá trị được ánh xạ có cùng khóa.

Container liên kết không có thứ tự

Container liên kết không có thứ tự lưu trữ dữ liệu không theo thứ tự cụ thể nào, nhưng chúng cho phép thực hiện các thao tác chèn, xóa và tìm kiếm nhanh nhất trong số tất cả các loại container trong STL.

Có 4 loại container liên kết không có thứ tự trong C++ STL:

1.  [Tập hợp không có thứ tự](https://www.geeksforgeeks.org/cpp/unordered_set-in-cpp-stl/) : STL Tập hợp không có thứ tự lưu trữ các khóa duy nhất dưới dạng bảng băm. Thứ tự được sắp xếp ngẫu nhiên nhưng việc chèn, xóa và tìm kiếm diễn ra nhanh chóng.
2.  [Unordered Multiset](https://www.geeksforgeeks.org/cpp/cpp-unordered_multiset/) : STL Unordered Multiset hoạt động tương tự như một tập hợp không có thứ tự nhưng có thể lưu trữ nhiều bản sao của cùng một khóa.
3.  [Bản đồ không có thứ tự](https://www.geeksforgeeks.org/cpp/unordered_map-in-cpp-stl/) : Bản đồ không có thứ tự STL lưu trữ cặp khóa-giá trị trong bảng băm, trong đó khóa được băm để tìm vị trí lưu trữ.
4.  [Unordered Multimap](https://www.geeksforgeeks.org/cpp/unordered_multimap-and-its-application/) : STL Unordered Multimap container tương tự như unordered map, nhưng nó cho phép nhiều giá trị được ánh xạ tới cùng một khóa.

Thuật toán

[Thuật toán STL](https://www.geeksforgeeks.org/cpp/c-magicians-stl-algorithms/) cung cấp một loạt các hàm để thực hiện các thao tác phổ biến trên dữ liệu (chủ yếu là container). Các hàm này triển khai phiên bản hiệu quả nhất của thuật toán cho các tác vụ như sắp xếp, tìm kiếm, sửa đổi và thao tác dữ liệu trong container, v.v. Tất cả các thuật toán STL đều được định nghĩa bên trong tệp tiêu đề <algorithm> và <numeric> .

Các hàm tử

[Functor](https://www.geeksforgeeks.org/cpp/functors-in-cpp/)  là các đối tượng có thể được xử lý như thể chúng là một hàm. Functor thường được sử dụng cùng với các thuật toán STL. Nó nạp  chồng toán tử gọi hàm  và cho phép chúng ta sử dụng một đối tượng như một hàm. Có nhiều functor được định nghĩa sẵn trong C++ STL, được định nghĩa bên trong tệp tiêu đề <functional> . ()

Functor có thể được phân loại thành nhiều loại dựa trên loại toán tử mà chúng thực hiện:

1.  Các hàm tử số học
2.  Hàm tử quan hệ
3.  Các hàm tử logic
4.  Các hàm tử bitwise

Các hàm tử số học

-   plus - Trả về tổng của hai tham số.
-   trừ - Trả về hiệu số của hai tham số.
-   nhân - Trả về tích của hai tham số.
-   chia - Trả về kết quả sau khi chia hai tham số.
-   môđun - Trả về phần dư sau khi chia hai tham số.
-   phủ định - Trả về giá trị phủ định của một tham số.

Hàm tử quan hệ

-   equal_to - Trả về true nếu hai tham số bằng nhau.
-   not_equal_to - Trả về true nếu hai tham số không bằng nhau.
-   lớn hơn - Trả về true nếu tham số đầu tiên lớn hơn tham số thứ hai.
-   greater_equal - Trả về true nếu tham số đầu tiên lớn hơn hoặc bằng tham số thứ hai.
-   less - Trả về true nếu tham số đầu tiên nhỏ hơn tham số thứ hai.
-   less_equal - Trả về true nếu tham số đầu tiên nhỏ hơn hoặc bằng tham số thứ hai.

Các hàm tử logic

-   logical_and - Trả về kết quả của phép toán logic AND của hai tham số.
-   logical_or - Trả về kết quả của phép toán OR logic của hai tham số.
-   logical_not - Trả về kết quả của phép toán logic NOT của các tham số.

Các hàm tử bitwise

-   bit_and - Trả về kết quả của phép toán Bitwise AND của hai tham số.
-   bit_or - Trả về kết quả của phép toán Bitwise OR của hai tham số.
-   bit_xor - Trả về kết quả của phép toán XOR bit của hai tham số.
