## // khởi tạo

```python
// string constructor
#include <iostream>
#include <string>

int main () {
    std::string s0 ("Initial string");

    // tạo string rỗng
    std::string s1;

    // s2 copy từ s1
    std::string s2 (s0);

    // tạo xâu con lấy từ s0, bắt đầu ở vị trí s0[8], kéo dài thêm 3 ký tự
    std::string s3 (s0, 8, 3);

    // tạo xâu con từ literal "A character sequence"
    std::string s4 ("A character sequence");

    // tạo xâu con từ literal "A character sequence", bắt đầu từ 0; kết thúc ở vị trí s4[11]
    std::string s5 ("Another character sequence", 12);
    
    // tạo xâu con gồm 10 số x
    std::string s6a (10, 'x');

    // tạo xâu con gồm 10 ký tự *, ký tự * có mã ascii là 42
    std::string s6b (10, 42);

    // tạo xâu con bắt đầu từ s0[0] tới s0[7]
    std::string s7 (s0.begin(), s0.begin()+7);

    return 0;
}
```

```python
s0: Initial string
s1: 
s2: Initial string
s3: str
s4: A character sequence
s5: Another char
s6a: xxxxxxxxxx
s6b: **********
s7: Initial
```

## // iterator

```python
// iterator : con chạy, có thể sửa
    // s1.begin() = đầu 
    // s1.end() = s1.begin()+ s1.length()= cuối

// reserve iterator : con chạy ngược, có thể sửa
    // s1.rbegin() = cuối
    // s1.rend() = s1.begin()+ s1.length()= đầu

// c iterator : con chạy, không thể sửa
    // s1.cbegin() = cuối
    // s1.cend() = s1.begin()+ s1.length()= đầu

// c r iterator : con chạy ngược, không thể sửa
    // s1.crbegin() = cuối
    // s1.crend() = s1.begin()+ s1.length()= đầu
```

```python
string str ("Test string");
for ( string::iterator it=str.begin(); it!=str.end(); ++it) {
    cout << *it;
}
cout << '\n';
```
